```python
import numpy as np
nb118_emitters = np.loadtxt('emitters in primer')
```

```python
#Creating lists to append to for each emission line group
RA_Halpha = []
DEC_Halpha = []
classic_ID_Halpha = []
z_lephare_Halpha = []
z_eazy_Halpha = []

RA_Hbeta_O3 = []
DEC_Hbeta_O3 = []
classic_ID_Hbeta_O3 = []
z_lephare_Hbeta_O3 = []
z_eazy_Hbeta_O3 = []


RA_O2 = []
DEC_O2 = []
classic_ID_O2 = []
z_lephare_O2 = []
z_eazy_O2 = []

RA_other = []
DEC_other = []
classic_ID_other = []
z_lephare_other = []
z_eazy_other = []

#Sorting the galaxies by redshift according to Eq. 4.1 in the thesis
for i in range (len(nb118_emitters[0])):
    RA_coor = nb118_emitters[0][i]
    DEC_coor = nb118_emitters[1][i]
    classic_ID = nb118_emitters[2][i]
    z_phot_lephare = nb118_emitters[3][i]
    z_phot_eazy = nb118_emitters[4][i]
    
    if 0.65 < z_phot_eazy < 0.95 and 0.65 < z_phot_lephare < 0.95:
        RA_Halpha.append(RA_coor)
        DEC_Halpha.append(DEC_coor)
        classic_ID_Halpha.append(classic_ID)
        z_lephare_Halpha.append(z_phot_lephare)
        z_eazy_Halpha.append(z_phot_eazy)
        
    elif 1.25 < z_phot_eazy < 1.55 and 1.25 < z_phot_lephare < 1.55:
        RA_Hbeta_O3.append(RA_coor)
        DEC_Hbeta_O3.append(DEC_coor)
        classic_ID_Hbeta_O3.append(classic_ID)
        z_lephare_Hbeta_O3.append(z_phot_lephare)
        z_eazy_Hbeta_O3.append(z_phot_eazy)
        
    elif 2.05 < z_phot_eazy < 2.35 and 2.05 < z_phot_lephare < 2.35:
        RA_O2.append(RA_coor)
        DEC_O2.append(DEC_coor)
        classic_ID_O2.append(classic_ID)
        z_lephare_O2.append(z_phot_lephare)
        z_eazy_O2.append(z_phot_eazy)
    
    else:
        RA_other.append(RA_coor)
        DEC_other.append(DEC_coor)
        classic_ID_other.append(classic_ID)
        z_lephare_other.append(z_phot_lephare)
        z_eazy_other.append(z_phot_eazy)

#converting Classic ID's to strings        
classic_ID_Halpha = ["%.0f" % number for number in classic_ID_Halpha]
classic_ID_Hbeta_O3 = ["%.0f" % number for number in classic_ID_Hbeta_O3]
classic_ID_O2 = ["%.0f" % number for number in classic_ID_O2]
classic_ID_other = ["%.0f" % number for number in classic_ID_other]

#Saving the sorted objects in external text files
np.savetxt('H-alpha emitters', np.array([RA_Halpha, DEC_Halpha, classic_ID_Halpha, z_lephare_Halpha, z_eazy_Halpha]), delimiter='\t', fmt="%s")
np.savetxt('H-beta and O3 emitters', np.array([RA_Hbeta_O3, DEC_Hbeta_O3, classic_ID_Hbeta_O3, z_lephare_Hbeta_O3, z_eazy_Hbeta_O3]), delimiter='\t', fmt="%s")
np.savetxt('O2 emitters', np.array([RA_O2, DEC_O2, classic_ID_O2, z_lephare_O2, z_eazy_O2]), delimiter='\t', fmt="%s")
np.savetxt('rest of emitters', np.array([RA_other, DEC_other, classic_ID_other, z_lephare_other, z_eazy_other]), delimiter='\t', fmt="%s")
```
